## Level 1 Task3 (Price Range distribution)

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"C:\Users\User\Downloads\Dataset.csv")
```

```python
dataset
```

```python
plt.figure(figsize=(4,8))
dataset['Price range'].value_counts().sort_index().plot(kind='bar',color='red')
plt.xlable('price range')
plt.ylable('Number of Restaurants')
plt.xtitle('Distribution of price Range Among Restaurants')
plt.xticks(rotation=0)
plt.show()
```

```python
total_restaurants=len(dataset)
Price_range_counts=dataset['Price range'].value_counts()
percentage_restaurants=(Price_range_counts/total_restaurants)*100
print("Percenntage of restaurants in each Price range category:")
print(percentage_restaurants)
```
